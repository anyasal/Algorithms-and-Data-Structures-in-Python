## Задача 2

Дан текстовый файл. В нем написан текст, разбитый на предложения. Перемешайте слова в каждом предложении текста и запишите полученный текст в новый файл.

````python
import random
import re
 
with open("input.txt", "r", encoding="utf-8") as file:
    text = file.read()
 
sentences = re.split(r'([.!?])', text)
result = ""
 
for i in range(0, len(sentences) - 1, 2):
    sentence = sentences[i].strip()
    punctuation = sentences[i + 1]
 
    words = sentence.split()
    random.shuffle(words)
 
    result += " ".join(words) + punctuation + " "
 
with open("output.txt", "w", encoding="utf-8") as file:
    file.write(result.strip())
 
print("результат записан в output.txt")
````

## Задача 3

В файле расположен русский текст. Откройте этот файл, преобразуйте текст в транслит и запишите в новый файл.

````python
translit_dict = {
    'а': 'a', 'б': 'b', 'в': 'v', 'г': 'g', 'д': 'd',
    'е': 'e', 'ё': 'yo', 'ж': 'zh', 'з': 'z', 'и': 'i',
    'й': 'y', 'к': 'k', 'л': 'l', 'м': 'm', 'н': 'n',
    'о': 'o', 'п': 'p', 'р': 'r', 'с': 's', 'т': 't',
    'у': 'u', 'ф': 'f', 'х': 'kh', 'ц': 'ts', 'ч': 'ch',
    'ш': 'sh', 'щ': 'shch', 'ъ': '', 'ы': 'y', 'ь': '',
    'э': 'e', 'ю': 'yu', 'я': 'ya'
}
 
with open("russian.txt", "r", encoding="utf-8") as file:
    text = file.read()
 
result = ""
 
for char in text:
    lower_char = char.lower()
    if lower_char in translit_dict:
        translit_char = translit_dict[lower_char]
        if char.isupper():
            translit_char = translit_char.capitalize()
        result += translit_char
    else:
        result += char
 
with open("translit.txt", "w", encoding="utf-8") as file:
    file.write(result)
 
print("транслит записан в translit.txt")
````

## Задача 4

Дан текстовый файл. Получите количество букв в нем (пробелы, знаки препинания и специальные символы не учитываются).

````python
count = 0
 
with open("text.txt", "r", encoding="utf-8") as file:
    text = file.read()
 
    for char in text:
        if char.isalpha():
            count += 1
 
print("количество букв:", count)
````

Примеры выполнения программы:
```
Первое значение: 10
Второе значение: 50
Результат: 60
```

```
Первое значение: hello
Второе значение: world
Результат: helloworld
```






````python
first = input()
second = input()
 
try:
    num1 = float(first)
    num2 = float(second)
 
    result = num1 + num2
 
    if result.is_integer():
        result = int(result)
 
    print("Результат:", result)
 
except ValueError:
    print("Результат:", first + second)
````
